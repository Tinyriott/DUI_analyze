```python
from google.oauth2 import service_account
from google.cloud import bigquery
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import font_manager,rc

plt.rc("font", family = "Malgun gothic")
```

```python
credentials = service_account.Credentials.from_service_account_file("./teamproject_key.json")
```

```python
client = bigquery.Client(
    credentials = credentials,
    project = credentials.project_id
)
```

```python
m_dui = """
SELECT
    FORMAT_DATE('%Y', `teamproject-428905.kim.dui_month`.`연도`) AS date,
    `1월`,
    `2월`,
    `3월`,
    `4월`,
    `5월`,
    `6월`,
    `7월`,
    `8월`,
    `9월`,
    `10월`,
    `11월`,
    `12월`
FROM
    `teamproject-428905.kim.dui_month`
"""
```

```python
m_dui_df = client.query(m_dui).to_dataframe()
```

```python
m_dui_df = m_dui_df.set_index("date")
```

```python
avg_m_dui = """
SELECT
    ROUND(AVG(`1월`)) AS `1월`,
    ROUND(AVG(`2월`)) AS `2월`,
    ROUND(AVG(`3월`)) AS `3월`,
    ROUND(AVG(`4월`)) AS `4월`,
    ROUND(AVG(`5월`)) AS `5월`,
    ROUND(AVG(`6월`)) AS `6월`,
    ROUND(AVG(`7월`)) AS `7월`,
    ROUND(AVG(`8월`)) AS `8월`,
    ROUND(AVG(`9월`)) AS `9월`,
    ROUND(AVG(`10월`)) AS `10월`,
    ROUND(AVG(`11월`)) AS `11월`,
    ROUND(AVG(`12월`)) AS `12월`
FROM
    `teamproject-428905.kim.dui_month`
"""
```

```python
avg_m_dui_df = client.query(avg_m_dui).to_dataframe()
```

```python
avg_m_dui_df = avg_m_dui_df.transpose()
```

```python
avg_m_dui_df.columns = ['평균값']
```
