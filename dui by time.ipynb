```python
from google.oauth2 import service_account
from google.cloud import bigquery
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import font_manager,rc

plt.rc("font", family = "Malgun gothic")
```

```python
credentials = service_account.Credentials.from_service_account_file("./teamproject_key.json")
```

```python
client = bigquery.Client(
    credentials = credentials,
    project = credentials.project_id
)
```

```python
time = """
SELECT
    *
FROM
    `teamproject-428905.kim.dui_accident_time`
"""
```

```python
time_df = client.query(time).to_dataframe()
```

```python
time_df = time_df.set_index("시간")
```

```python
time_df
```

```python
plt.style.use ('ggplot')
time_df.plot(kind = "bar")
_ = plt.xticks(size = 8, rotation=45, ha='right')
```

```python
sorted_time_df = time_df.sort_values(by='사고건수', ascending=False)
```

```python
plt.style.use ('ggplot')
sorted_time_df.plot(kind = "bar")
_ = plt.xticks(size = 8, rotation=45, ha='right')
```
