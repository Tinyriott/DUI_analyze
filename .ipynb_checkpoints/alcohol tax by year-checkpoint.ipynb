```python
from google.oauth2 import service_account
from google.cloud import bigquery
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import font_manager,rc

plt.rc("font", family = "Malgun gothic")
```

```python
credentials = service_account.Credentials.from_service_account_file("./teamproject_key.json")
```

```python
client = bigquery.Client(
    credentials = credentials,
    project = credentials.project_id
)
```

```python
tax = """
SELECT
    FORMAT_DATE('%Y', `teamproject-428905.kim.tax_year`.`연도`) AS `연도`,
    `teamproject-428905.kim.tax_year`.`주류세`
FROM
    `teamproject-428905.kim.tax_year`
"""
```

```python
tax_df = client.query(tax).to_dataframe()
```

```python
tax_df = tax_df.set_index("연도")
```

```python
tax_df
```

```python
plt.style.use ('ggplot')
tax_df.plot(marker = "o", linestyle='-', figsize = (12,6))
plt.title("< 주류세 변화 추이 >")
plt.xlabel("연도")
plt.ylabel("총액")
```
